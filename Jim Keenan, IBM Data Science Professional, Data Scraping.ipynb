### The main thing for getting data - in this case a table - From a website using public data. BeautifulSoup is cool

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
postal_r=requests.get(url)
postal_soup=BeautifulSoup.BeautifulSoup(postal_r.text,'html.parser')
postal_soup

data = []
table = postal_soup.find('table', attrs={'class':'wikitable sortable'})
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
     cols = row.find_all('td')
     cols = [ele.text.strip() for ele in cols]
     data.append([ele for ele in cols if ele]) # Get rid of empty values

type(data)
df=pd.DataFrame(data)
df.columns=['Postcode','Borough','Neighborhood']
for row in df.index:
    if df['Borough'][row]=='Not assigned':
        df.drop(row,inplace=True)
   
df2=df
```

### Removed all the 'Non assigned' candidates

```python
for row in df2.index:
    if df2['Neighborhood'][row]=='Not assigned':
        df2['Neighborhood'][row]=df2['Borough'][row]
df2
```
